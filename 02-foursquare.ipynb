<h1>Folium Group Map with all the locations<h1>

```python
sanfran_map = folium.Map(location= [37.788482,-122.409173], zoom_start= 15, tiles="cartodbpositron")

new_company_group = folium.FeatureGroup(name = "New Company Location")
new = Marker(location= [37.788482,-122.409173], icon=folium.Icon(color='lightblue', icon='institution', prefix='fa'), tooltip="My Company")
new.add_to(new_company_group)
new_company_group.add_to(sanfran_map)




starbucks_group = folium.FeatureGroup(name = "Starbucks")
school_group = folium.FeatureGroup(name = "Schools")
vegan_group = folium.FeatureGroup(name = "Vegan Restaurnats")
basketball_group = folium.FeatureGroup(name = "Basketball Arenas")
night_clubs_group = folium.FeatureGroup(name = "Night Clubs")
dobby_group = folium.FeatureGroup(name = "Dog Grooming")

for index, row in starbucks_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='green', icon='coffee', prefix='fa'), tooltip=row["distance"]).add_to(starbucks_group)
starbucks_group.add_to(sanfran_map)    

for index, row in vegan_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='black', icon='leaf', prefix='fa'), tooltip=row["distance"]).add_to(vegan_group)
vegan_group.add_to(sanfran_map)    

for index, row in school_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='red', icon='book', prefix='fa'), tooltip=row["distance"]).add_to(school_group)
school_group.add_to(sanfran_map)

    
for index, row in night_clubs_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='purple', icon='glass', prefix='fa'), tooltip=row["distance"]).add_to(night_clubs_group)
night_clubs_group.add_to(sanfran_map)

    
for index, row in basketball_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='blue', icon='futbol-o', prefix='fa'), tooltip=row["distance"]).add_to(basketball_group)
basketball_group.add_to(sanfran_map)

    
for index, row in dobby_df.iterrows():
    Marker(location = [row["latitude"], row["longitude"]], icon=folium.Icon(color='orange', icon='gitlab', prefix='fa'), tooltip=row["distance"]).add_to(dobby_group)
dobby_group.add_to(sanfran_map)




folium.LayerControl(collapsed=False).add_to(sanfran_map)
```
